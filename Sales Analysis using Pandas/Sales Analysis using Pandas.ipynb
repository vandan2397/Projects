```python
duplicate = Data[Data.duplicated('Order ID',keep=False)]
```

```python
duplicate.sort_values(by='Order ID', inplace=True)
```

```python
duplicate.head(10)
```

```python
duplicate['Grouped'] = duplicate.groupby(['Order ID'])['Product'].transform(lambda x: ','.join(x))
```

```python
duplicate.head()
```

```python
#remove duplicates
unique_data = duplicate[['Order ID','Grouped']].drop_duplicates()
unique_data
```

```python
# More often purchased products
from itertools import combinations
from collections import Counter


counter = Counter()

for row in unique_data['Grouped']:
    row_list = row.split(',')
    counter.update(Counter(combinations(row_list,2)))
    
counter.most_common(10)
```

```python
for key, value in counter.most_common():
    print(key, value)
```
